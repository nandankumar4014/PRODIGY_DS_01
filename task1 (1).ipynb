```python
employees_by_education = df['Education Level'].value_counts()

plt.figure(figsize=(10, 6))
employees_by_education.plot(kind='bar', color='orange')
plt.title('Number of Employees by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
avg_salary_by_gender = df.groupby('Gender')['Salary'].mean()
avg_salary_by_gender.plot(kind='bar', color=['lightblue', 'lightgreen'])
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
avg_salary_by_education = df.groupby('Education Level')['Salary'].mean()

avg_salary_by_education.plot(kind='bar', color='skyblue', figsize=(8, 5))
plt.title('Average Salary by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Salary')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Histogram for the distribution of salaries
axes[0].hist(df['Salary'], bins=10, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Salaries')
axes[0].set_xlabel('Salary')
axes[0].set_ylabel('Frequency')

# Histogram for the distribution of ages
axes[1].hist(df['Age'], bins=10, color='lightgreen', edgecolor='black')
axes[1].set_title('Distribution of Ages')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Years of Experience'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Distribution of Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
